## Questions (8+)

### Quick view of some of the top tracks based on popularity.

1. Sort all the tracks by popularity in decreasing order.

```python
top_tracks = df.sort_values(['popularity'], ascending=False)
top_tracks
```

### Now section out the ones that have a popularity of 50 or higher

```python
top_fifty = df[df.popularity >= 50]
top_fifty
```

Now that we have it sectioned we are going to do some basic checks on that information.

2. What is the most common genre (what appears the most)

```python
top_fifty.track_genre.value_counts()
```

3. What is the average danceability of these tracks?

```python
top_fifty.groupby('track_genre').danceability.mean()
```

4. What is the max,min, and average duration of these tracks?

```python
answer = pd.DataFrame(top_fifty.groupby('track_genre').duration_ms.agg(["max","min","mean"]))
answer
```

5. What is the average instrumentalness,tempo, and energy of these tracks?

```python
top_fifty.groupby('track_genre')['instrumentalness','tempo','energy'].mean()
```

## Visualizations (4+)

1. Make a line chart to see the correlation between danceability and energy when the popularity levels change.
(Put Danceability and energy on the y-axis and popularity on the x-axis)

```python
sns.lineplot(x=top_fifty.popularity, y=top_fifty.danceability, color='Purple', ci=None, label="Danceability")
sns.lineplot(x=top_fifty.popularity, y=top_fifty.energy, color='Green', ci=None, label="Energy")
plt.xlabel("Popularity")
plt.ylabel("Danceability and Energy")
plt.title("Danceability vs. Energy (Top 50)",fontsize=16)
```

```python
sns.lineplot(x=bottom_fifty.popularity, y=bottom_fifty.danceability, color='Blue', ci=None, label="Danceability")
sns.lineplot(x=bottom_fifty.popularity, y=bottom_fifty.energy, color='Orange', ci=None, label="Energy")
plt.xlabel("Popularity")
plt.ylabel("Danceability and Energy")
plt.title("Danceability vs. Energy (Bottom 50)",fontsize=16)
```

2. Make a line chart to see the correlation between speechiness and explicit when the popularity levels change.
(Do this graph like you did the first)

```python
sns.lineplot(x=top_fifty.popularity, y=top_fifty.speechiness, color='Blue', ci=None, label="Speechiness")
sns.lineplot(x=top_fifty.popularity, y=top_fifty.explicit, color='Orange', ci=None, label="Explicit")
plt.xlabel("Popularity")
plt.ylabel("Speechiness and Explicit")
plt.title("Speechiness vs. Explicit (Top 50)",fontsize=16)
```

```python
sns.lineplot(x=bottom_fifty.popularity, y=bottom_fifty.speechiness, color='Green', ci=None, label='Speechiness')
sns.lineplot(x=bottom_fifty.popularity, y=bottom_fifty.explicit, color='Red', ci=None, label="Explicit")
plt.xlabel("Popularity")
plt.ylabel("Speechiness and Explicit")
plt.title("Speechiness vs. Explicit (Bottom 50)",fontsize=16)
```

3. Plot the average duration vs the popularity for each genre on a scatter plot (with popularity on the x axis and average duration on the y-axis).

```python
plt.figure(figsize=(22,12))
sns.set_style('darkgrid')
sns.scatterplot(
    x=top_fifty.popularity,
    y=top_fifty.duration_ms,
    data=top_fifty.track_genre,
    hue=top_fifty.track_genre,
    palette='bright'
)
plt.xlabel("Popularity")
plt.ylabel("Duration")
plt.title("Avg Popularity vs. Avg Duration (Top 50)",fontsize=16)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

```python
plt.figure(figsize=(22,12))
sns.set_style('darkgrid')
sns.scatterplot(
    x=bottom_fifty.popularity,
    y=bottom_fifty.duration_ms,
    data=bottom_fifty.track_genre,
    hue=bottom_fifty.track_genre,
    palette='bright'
)
plt.xlabel("Popularity")
plt.ylabel("Duration")
plt.title("Avg Popularity vs. Avg Duration (Top 50)",fontsize=16)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

4. Plot the popularity of each genre in a horizontal bar chart(with popularity on the horizontal axis and genre on the vertical axis). (The entire dataframe will be used)

```python
plt.figure(figsize=(34, 24))
sns.set_style('white')
ax = sns.barplot(
    y= df.track_genre,
    x= df.popularity,
    ci=None
)

for i in ax.containers:
    ax.bar_label(i,)

plt.title("Avg Popularity for each Genre", fontsize=16)
plt.xlabel("Popularity",fontsize=13)
plt.ylabel("Genres",fontsize=13)
```
