## Questions (8+)

### Quick view of some of the top tracks based on popularity.

1. Sort all the tracks by popularity in decreasing order.

```python
top_tracks = df.sort_values(['popularity'], ascending=False)
top_tracks
```

### Now section out the ones that have a popularity of 50 or higher

```python
top_fifty = df[df.popularity >= 50]
top_fifty
```

Now that we have it sectioned we are going to do some basic checks on that information.

2. What is the most common genre (what appears the most)

```python
top_fifty.track_genre.value_counts()
```

3. What is the average danceability of these tracks?

```python
top_fifty.groupby('track_genre').danceability.mean()
```

4. What is the max,min, and average duration of these tracks?

```python
answer = pd.DataFrame(top_fifty.groupby('track_genre').duration_ms.agg(["max","min","mean"]))
answer
```

5. What is the average instrumentalness,tempo, and energy of these tracks?

```python
top_fifty.groupby('track_genre')['instrumentalness','tempo','energy'].mean()
```
